```python
# problem-2
review = input("Enter your review:")

review = review.lower()

words = review.split()

word_count = {}

for w in words:
    if w in word_count:
        word_count[w] += 1
    else:
        word_count[w] = 1

print(word_count)
```

```python
# Problem-3

password = input("Enter your password: ")

upper = False
lower = False
digit = False
special = False

special_chars = "!@#$%^&*()_+-=<>?/{}[]|"

for c in password:
    if c.isupper():
        upper = True
    elif c.islower():   
        lower = True
    elif c.isdigit():
        digit = True
    elif c in special_chars:
        special = True    

if upper and lower and digit and special:
    print("Strong Password")
else:
    print("Weak Password")
```

```python
# problem-4
import random
name=input("Enter full name: ")
name=name.lower()
first=""
last=""
found_space=""

for ch in name:
    if ch==" ":
        found_space=True
        continue
    if not found_space:
        first+=ch
    else:
        last+=ch

first3=""
count=0
for ch in first:
    if count==3:
        break
    first3+=ch
    count+=1

last3=""
length=0
for ch in last:
    length+=1
for i in range(length-3,length):
    last3+=last[i]

num=random.randint(10,99)

username=first3+last3+str(num)

print(username)
```

```python
# problem-5

msg = input("Enter your message: ")


char_count = 0
for c in msg:
    char_count += 1


word_count = 0
word = False

for c in msg:
    if c != " " and not word:
        word_count += 1
        word = True
    elif c == " ":
        word = False


if char_count > 160:
    status = "Exceeds Limit"
else:
    status = "Within Limit"

print("Words =", word_count)
print("Characters =", char_count)
print("Status =", status)
```

```python
# problem-6

names = input("Enter names (comma separated): ")
names=names.split(",")
unique_names = []

for name in names:
    name = name.strip()  
    if name not in unique_names:
        unique_names.append(name)


result = ", ".join(unique_names)
print(result)
```

```python
# problem-9
post = input("Give your post: ")
hashtags = 0
mentions = 0

for c in post:
    if c == "#":
        hashtags += 1
    elif c == "@":
        mentions += 1

print("Hashtags =", hashtags)
print("Mentions =", mentions)
```

```python
# problem-10
sentence = input("Give your sentence: ")

words = sentence.split()
reversed_words = []

for word in words:
    rev = ""
    for c in word:
        rev = c + rev 
    reversed_words.append(rev)

result = " ".join(reversed_words)
print(result)
```
