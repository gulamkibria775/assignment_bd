```python
num=int(input("enter you number (0-100):"))
if(90<=num<100):
    print("A+")
elif(80<=num<=89):
    print("A")  
elif(70<=num<=79):
    print("B")
elif(60<=num<=69):
    print("C")
else:
    print("Fail")   
```

```python
color=input("enter a color:")
color=color.lower()
if(color=="red"):
    print("Stop")
elif(color=="yellow"):
    print("Ready to go")
elif(color=="green"):
    print("Go")
else:
    print("Invalid color")
```

```python
# Assignment 8: Leap Year Checker
```

```python
year=int(input("give me a year to check leapyear:"))
if(year%400==0):
    print(f"{year} is leap year")
elif(year%4==0 and year%100!=0):
    print(f"{year} is leap year")
else:
    print(f"{year} is not leap year")
```
