```python
# problem-1

def calculate_inventory_summary(prices):
    total_cost=sum(prices)
    if(prices):
        average_cost=total_cost/len(prices)
    else:
        average_cost=0
    return total_cost,average_cost

prices = [120, 250, 399, 150]
total,average=calculate_inventory_summary(prices)
print(f"total price = ",total)
print(f"average price =",average)
```

```python
# problem-2

def counter_unique_visitor(visitor_list):
    unique_visitor=set(visitor_list)
    return unique_visitor,len(unique_visitor)
visitor_list=[101, 205, 101, 310, 205, 550]

unique_visitor_list,unique_visitor_num=counter_unique_visitor(visitor_list)
print(f"{unique_visitor_num} (The unique IDs are ",unique_visitor_list,")" )
```

```python
# problem-3

def count_categories(products):
    category_count = {} 
    for product in products:
        if product in category_count:
            category_count[product] += 1
        else:
            category_count[product] = 1
    return category_count

products = ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]
print(count_categories(products))

```

```python
# problem-4

def analyze_temperatures(temp_list):
    high_list = [temp for temp in temp_list if temp > 38]      
    normal_list = [temp for temp in temp_list if temp <= 38]   
    return high_list, normal_list

temps = [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]
high, normal = analyze_temperatures(temps)

print("High List (>38°C):", high)
print("Normal List (≤38°C):", normal)
print("Full Output Tuple:", (high, normal))
```

```python
# problem-5

def get_price(menu,item):
    if item in menu:
        return[item]
    else:
        return "Item not found"
    
menu = {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
print(get_price(menu, "Muffin")) 
print(get_price(menu, "Donut"))   
```

```python
# problem-6

def calculate_grade(score):
    if score>=80:
        return "A+"
    elif score>=70:
        return "A"
    elif score >=60:
        return "A-"
    else:
        return "F"
    
def grade_students(student):
    grades={}
    for student,score in student.items():
        grades[student]=calculate_grade(score)
    return grades

scores = {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}
grades = grade_students(scores)
print(grades)
```

```python
# problem-7

def apply_discounts(products):
    discounted_prices = []
    for name, price, discount in products:
        updated_price = price * (1 - discount / 100)  
        discounted_prices.append(round(updated_price, 2))  
    return discounted_prices


products = [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]
print(apply_discounts(products))
```

```python
# problem-8

def generate_hashtags(sentence):
    words = sentence.split()                 
    hashtags = ['#' + word.lower() for word in words]  
    return hashtags

sentence = "AI Data Science"
print(generate_hashtags(sentence))
```

```python
# problem-9

CREDENTIALS = {
    "alex_s": "securepwd1",
    "maria_t": "password101",
    "john_d": "johndoe123"
}

def login(username, password):
    if username in CREDENTIALS and CREDENTIALS[username] == password:
        return "Login Successful"
    else:
        return "Invalid Credentials"

print(login("maria_t", "password101"))  
print(login("alex_s", "securepwd12"))    
print(login("peter_z", "any_password")) 
```

```python
# problem-10

def calculate_total(cart, prices):
    total=0
    for item in cart:
        if item in prices:
            total+=prices[item]
    return total

cart = ["Milk", "Bread", "Milk", "Cheese", "Bread"]
prices = {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}

print(calculate_total(cart, prices)) 
```
